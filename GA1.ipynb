# Table 1

```python
import pandas as pd

file = "Table1/table-1-1.xlsx"

# headers on row 5
df = pd.read_excel(file, header=5)
print("\033[1m" + "Table 1. Educational Attainment of the Population 18 Years and Over, by Age, Sex, Race, and Hispanic Origin: 2022 (All Races)\033[0m")

# ":," means select all rows, 0 is for the column
# this just gets rid of the tabs in the first column because they look weird when doing df.head()
# turning regex on means it'll find the \t anywhere in the string
df.iloc[:,0] = df.iloc[:,0].replace("\t", "", regex=True) 

# These rows were just descriptions of the rows underneath
# 0 said Both Sexes, 15 said Male, 30 said Female
df = df.drop([0, 15, 30])

all_both_sex = df.iloc[0:14].copy()
all_male = df.iloc[14:28].copy()     
all_female = df.iloc[28:42].copy()   

all_both_sex.rename(columns={df.columns[0]: 'Both Sex'}, inplace=True)
all_male.rename(columns={df.columns[0]: 'Male'}, inplace=True)
all_female.rename(columns={df.columns[0]: 'Female'}, inplace=True)

all_female

```
